```python
from sklearn.impute import SimpleImputer
```

```python
num_imputer = SimpleImputer(strategy='median')
df[['Age','Blood Pressure','Cholesterol Level','BMI','Sleep Hours','Triglyceride Level','Fasting Blood Sugar','CRP Level','Homocysteine Level']]=num_imputer.fit_transform(df[['Age','Blood Pressure','Cholesterol Level','BMI','Sleep Hours','Triglyceride Level','Fasting Blood Sugar','CRP Level','Homocysteine Level']])
```

```python
cat_imputer = SimpleImputer(strategy='most_frequent')
df[['Gender','Exercise Habits','Smoking','Family Heart Disease','Diabetes','High Blood Pressure','Low HDL Cholesterol','High LDL Cholesterol','Alcohol Consumption','Stress Level','Sugar Consumption']]=cat_imputer.fit_transform(df[['Gender','Exercise Habits','Smoking','Family Heart Disease','Diabetes','High Blood Pressure','Low HDL Cholesterol','High LDL Cholesterol','Alcohol Consumption','Stress Level','Sugar Consumption']])
```

```python
df.isna().sum()
```

```python
df['Heart Disease Status'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
cols = ['Gender','Exercise Habits','Smoking','Family Heart Disease','Diabetes','High Blood Pressure','Low HDL Cholesterol','High LDL Cholesterol','Alcohol Consumption','Stress Level','Sugar Consumption','Heart Disease Status']
for col in cols:
    df[col]=encode.fit_transform(df[col])
```

```python
a =['Age','Gender','Blood Pressure','Cholesterol Level','Exercise Habits','Smoking','Family Heart Disease','Diabetes','BMI','High Blood Pressure','High LDL Cholesterol','Alcohol Consumption','Stress Level','Sleep Hours','Sugar Consumption','Triglyceride Level','Fasting','Blood Sugar','CRP Level','Homocysteine Level']
b = 'Heart Disease Status'
for i in a:
    sns.lmplot(x=i,y=b,data=df)
    plt.title(f"{i} Vs {b}")
    plt.show()
```

```python
for col in a:
    plt.figure(figsize=(5, 4))
    sns.distplot(df[col],hist=False)
    plt.title(f"Normality Check: {col}")
    plt.show()
```

```python
df.corr()
```

```python
x = df.iloc[:,:-1]
x.head()
```

```python
y = df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=7)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
rfc_train_pred = rfc.predict(x_train)
rfc_train_pred
```

```python
rfc_test_pred = rfc.predict(x_test)
rfc_test_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
rfc_train_accuracy = accuracy_score(y_train,rfc_train_pred)
rfc_train_accuracy
```

```python
rfc_test_accuracy = accuracy_score(y_test,rfc_test_pred)
rfc_test_accuracy
```

```python
print(classification_report(y_train,rfc_train_pred))
```

```python
print(classification_report(y_test,rfc_test_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
```

```python
ada = AdaBoostClassifier()
ada.fit(x_train,y_train)
```

```python
ada_train_pred = ada.predict(x_train)
ada_train_pred
```

```python
ada_train_accuracy = accuracy_score(y_train,ada_train_pred)
ada_train_accuracy
```

```python
ada_test_pred = ada.predict(x_test)
ada_test_accuracy = accuracy_score(y_test,ada_test_pred)
ada_test_accuracy
```

```python
print(classification_report(y_train,ada_train_pred))
```

```python
print(classification_report(y_test,ada_test_pred))
```

```python
gbc = GradientBoostingClassifier()
gbc.fit(x_train,y_train)
```

```python
gbc_train_pred = gbc.predict(x_train)
gbc_train_accuracy = accuracy_score(y_train,gbc_train_pred)
gbc_train_accuracy
```

```python
gbc_test_pred = gbc.predict(x_test)
gbc_test_accuracy = accuracy_score(y_test,gbc_test_pred)
gbc_test_accuracy
```

```python
print(classification_report(y_train,gbc_train_pred))
```

```python
print(classification_report(y_test,gbc_test_pred))
```
